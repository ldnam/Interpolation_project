```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
from numpy import sqrt,log,exp
from PIL import Image
```

```python
im= Image.open("python.JFIF")
```

```python
a= np.array(im)
a.shape
```

```python
def not_six_points(i,j):
    return not ((i in [0,450,900]) and (j in [0,450]))
def dis(i,j):
    return sqrt((i[0]-j[0])**2+(i[1]-j[1])**2)
```

```python
r=np.zeros((1189,669,6))
t=0
h=[[0,0],[0,450],[450,0],[450,450],[900,0],[900,450]]
s=np.array([a[0,0],a[450,0],a[0,450],a[450,450],a[0,900],a[450,900]])
s=np.transpose(s)
```

```python
for i in range(1189):
    for j in range(669):
        if not_six_points(i,j):
            a[j,i]=0
```

```python
for i in range(1189):
    for j in range(669):
        if not_six_points(i,j):
            r[i,j]=[dis([i,j],h[0]),dis([i,j],h[1]),dis([i,j],h[2]),dis([i,j],h[3]),dis([i,j],h[4]),dis([i,j],h[5])]
            mnpq=(1/(1+r[i,j]**2))
            t=sum(mnpq)
            a[j,i]=(s@mnpq)/t
```

```python
a=a.astype("uint8")
plt.imshow(a)
plt.axis("off")
plt.show()
```

```python
plt.imshow(a)
plt.axis("off")
plt.show()
```
