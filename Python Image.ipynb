```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sqrt,log,exp
from PIL import Image
```

```python
im= Image.open("C:\\Users\\Nam Le\\python.JPG")
```

```python
a= np.array(im)
a.shape
```

```python
def not_six_points(i,j):
    return not ((i in [0,2500,5000]) and (j in [1500,3000]))
def dis(i,j):
    return sqrt((i[0]-j[0])**2+(i[1]-j[1])**2)
```

```python
r=np.zeros((5472,3078,6))
t=0
h=[[0,1500],[0,3000],[2500,1500],[2500,3000],[5000,1500],[5000,3000]]
s=np.array([a[1500,0],a[3000,0],a[1500,2500],a[3000,2500],a[1500,5000],a[3000,5000]])
s=np.transpose(s)
```

```python
for i in range(5472):
    for j in range(3078):
        if not_six_points(i,j):
            a[j,i]=0
```

```python
for i in range(5472):
    for j in range(3078):
        if not_six_points(i,j):
            r[i,j]=[dis([i,j],h[0]),dis([i,j],h[1]),dis([i,j],h[2]),dis([i,j],h[3]),dis([i,j],h[4]),dis([i,j],h[5])]
for i in range(5472):
    for j in range(3078):
        if not_six_points(i,j):          
            mnpq=(1/(1+r[i,j]**2))
            t=sum(mnpq)
            a[j,i]=(s@mnpq)/t
```

```python
a=a.astype("uint8")
plt.imshow(a)
plt.axis("off")
plt.show()
```

```python
for i in range(5472):
    for j in range(3078):
        if not_six_points(i,j):
            mnpq=sqrt(100+(r[i,j]**2))
            t=sum(mnpq)
            a[j,i]=(s@mnpq)/t
```

```python
plt.imshow(a)
plt.axis("off")
plt.show()
```

```python
for i in range(5472):
    for j in range(3078):
        if not_six_points(i,j):
            mnpq=(r[i,j]**2)*log(r[i,j])
            t=sum(mnpq)
            a[j,i]=(s@mnpq)/t
```

```python
plt.imshow(a)
plt.axis("off")
plt.show()
```

```python
for i in range(5472):
    for j in range(3078):
        if not_six_points(i,j):
            mnpq=exp(-1/1000*(r[i,j]**2))
            t=sum(mnpq)
            a[j,i]=(s@mnpq)/t
```

```python
plt.imshow(a)
plt.axis("off")
plt.show()
```
